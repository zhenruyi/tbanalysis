```python
import numpy as np
import matplotlib.pyplot as plt

# 定义一个简单的目标函数（二次函数）
def objective_function(x):
    return x**2 + 5 * np.sin(x)

# 定义目标函数的导数（梯度）
def gradient(x):
    return 2 * x + 5 * np.cos(x)

# 梯度下降算法
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    history = [x]  # 记录每次迭代的参数值
    for _ in range(num_iterations):
        grad = gradient(x)  # 计算梯度
        x = x - learning_rate * grad  # 更新参数
        history.append(x)
    return x, history

# 可视化目标函数和梯度下降过程
def plot_objective_function_and_gradient_descent():
    # 定义x的范围
    x_values = np.linspace(-10, 10, 400)
    y_values = objective_function(x_values)
    
    # 绘制目标函数
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label=r"目标函数: $f(x) = x^2 + 5 \sin(x)$", color="blue")
    
    # 设置梯度下降的初始点和学习率
    starting_point = -8  # 初始点
    learning_rate = 0.1  # 学习率
    num_iterations = 50  # 迭代次数
    
    # 运行梯度下降
    final_x, history = gradient_descent(starting_point, learning_rate, num_iterations)
    
    # 绘制梯度下降的过程
    for i, x in enumerate(history):
        y = objective_function(x)
        plt.scatter(x, y, color="red", s=50)
        if i > 0:
            plt.plot([history[i-1], x], [objective_function(history[i-1]), y], color="red", linestyle="--")
        plt.pause(0.1)  # 动态展示
    
    # 添加标题和标签
    plt.title("目标函数与梯度下降过程")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()

# 运行可视化
plot_objective_function_and_gradient_descent()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 设置支持中文的字体
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
matplotlib.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 定义目标函数
def objective_function(x):
    return x**2 + 5 * np.sin(x)

# 定义目标函数的导数（梯度）
def gradient(x):
    return 2 * x + 5 * np.cos(x)

# 梯度下降算法
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    history = [x]  # 记录每次迭代的参数值
    for _ in range(num_iterations):
        grad = gradient(x)  # 计算梯度
        x = x - learning_rate * grad  # 更新参数
        history.append(x)
    return x, history

# 可视化目标函数和梯度下降过程
def plot_objective_function_and_gradient_descent():
    # 定义x的范围
    x_values = np.linspace(-10, 10, 400)
    y_values = objective_function(x_values)
    
    # 绘制目标函数
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label=r"目标函数: $f(x) = x^2 + 5 \sin(x)$", color="blue", linewidth=2)
    
    # 设置梯度下降的初始点和学习率
    starting_point = -8  # 初始点
    learning_rate = 0.1  # 学习率
    num_iterations = 20  # 迭代次数
    
    # 运行梯度下降
    final_x, history = gradient_descent(starting_point, learning_rate, num_iterations)
    
    # 绘制梯度下降的过程
    history_y = [objective_function(x) for x in history]  # 计算每次迭代的函数值
    plt.scatter(history, history_y, color="red", s=100, zorder=5, label="梯度下降路径")
    plt.plot(history, history_y, color="red", linestyle="--", linewidth=1.5, alpha=0.6)
    
    # 标记初始点和最终点
    plt.scatter(history[0], history_y[0], color="green", s=150, marker="*", zorder=10, label="初始点")
    plt.scatter(history[-1], history_y[-1], color="purple", s=150, marker="X", zorder=10, label="最终点")
    
    # 添加标题和标签
    plt.title("目标函数与梯度下降过程", fontsize=16)
    plt.xlabel("x", fontsize=14)
    plt.ylabel("f(x)", fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

# 运行可视化
plot_objective_function_and_gradient_descent()
```
