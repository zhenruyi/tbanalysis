```python
import numpy as np
import matplotlib.pyplot as plt

# 
def objective_function(x):
    return x**2 + 5 * np.sin(x)

# 
def gradient(x):
    return 2 * x + 5 * np.cos(x)

# 
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    history = [x]  # 
    for _ in range(num_iterations):
        grad = gradient(x)  # 
        x = x - learning_rate * grad  # 
        history.append(x)
    return x, history

# 
def plot_objective_function_and_gradient_descent():
    # x
    x_values = np.linspace(-10, 10, 400)
    y_values = objective_function(x_values)
    
    # 
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label=r": $f(x) = x^2 + 5 \sin(x)$", color="blue")
    
    # 
    starting_point = -8  # 
    learning_rate = 0.1  # 
    num_iterations = 50  # 
    
    # 
    final_x, history = gradient_descent(starting_point, learning_rate, num_iterations)
    
    # 
    for i, x in enumerate(history):
        y = objective_function(x)
        plt.scatter(x, y, color="red", s=50)
        if i > 0:
            plt.plot([history[i-1], x], [objective_function(history[i-1]), y], color="red", linestyle="--")
        plt.pause(0.1)  # 
    
    # 
    plt.title("")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()

# 
plot_objective_function_and_gradient_descent()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 
matplotlib.rcParams['axes.unicode_minus'] = False  # 

# 
def objective_function(x):
    return x**2 + 5 * np.sin(x)

# 
def gradient(x):
    return 2 * x + 5 * np.cos(x)

# 
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    history = [x]  # 
    for _ in range(num_iterations):
        grad = gradient(x)  # 
        x = x - learning_rate * grad  # 
        history.append(x)
    return x, history

# 
def plot_objective_function_and_gradient_descent():
    # x
    x_values = np.linspace(-10, 10, 400)
    y_values = objective_function(x_values)
    
    # 
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label=r": $f(x) = x^2 + 5 \sin(x)$", color="blue", linewidth=2)
    
    # 
    starting_point = -8  # 
    learning_rate = 0.1  # 
    num_iterations = 20  # 
    
    # 
    final_x, history = gradient_descent(starting_point, learning_rate, num_iterations)
    
    # 
    history_y = [objective_function(x) for x in history]  # 
    plt.scatter(history, history_y, color="red", s=100, zorder=5, label="")
    plt.plot(history, history_y, color="red", linestyle="--", linewidth=1.5, alpha=0.6)
    
    # 
    plt.scatter(history[0], history_y[0], color="green", s=150, marker="*", zorder=10, label="")
    plt.scatter(history[-1], history_y[-1], color="purple", s=150, marker="X", zorder=10, label="")
    
    # 
    plt.title("", fontsize=16)
    plt.xlabel("x", fontsize=14)
    plt.ylabel("f(x)", fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

# 
plot_objective_function_and_gradient_descent()
```
