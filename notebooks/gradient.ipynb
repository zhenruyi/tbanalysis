```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 定义损失函数（这里用一个简单的二次函数 y = x^2 + 5）
def loss_function(x):
    return x**2 + 5

# 2. 定义损失函数的梯度（导数）
def gradient(x):
    return 2 * x

# 3. 梯度下降算法
def gradient_descent(starting_point, learning_rate, num_iterations):
    # 记录每次迭代的参数和损失值
    history = []
    x = starting_point
    
    for i in range(num_iterations):
        # 计算梯度
        grad = gradient(x)
        # 更新参数
        x = x - learning_rate * grad
        # 记录当前参数和损失值
        history.append((x, loss_function(x)))
    
    return history

# 4. 设置初始参数
start_point = 10  # 起点
learning_rate = 0.1  # 学习率
num_iterations = 20  # 迭代次数

# 5. 运行梯度下降
history = gradient_descent(start_point, learning_rate, num_iterations)

# 6. 可视化结果
# 绘制损失函数曲线
x_values = np.linspace(-11, 11, 400)
y_values = loss_function(x_values)
plt.plot(x_values, y_values, label="Loss Function: $y = x^2 + 5$")

# 绘制梯度下降的过程
x_history = [h[0] for h in history]
y_history = [h[1] for h in history]
plt.scatter(x_history, y_history, color="red", label="Gradient Descent Steps")
plt.plot(x_history, y_history, color="red", linestyle="--")

# 标记起点和终点
plt.scatter(start_point, loss_function(start_point), color="green", label="Start Point")
plt.scatter(x_history[-1], y_history[-1], color="blue", label="End Point")

# 添加标签和图例
plt.xlabel("x")
plt.ylabel("Loss")
plt.title("Gradient Descent Visualization")
plt.legend()
plt.grid()
plt.show()
```
