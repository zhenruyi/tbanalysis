```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import euclidean, cityblock, cosine
from sklearn.metrics import jaccard_score

# Set random seed
np.random.seed(42)

# Generate example data
# Two 2D vectors
x = np.array([1, 2])
y = np.array([4, 6])

# Two sets (for Jaccard similarity)
set_x = np.array([1, 0, 1, 1, 0])  # Binary data
set_y = np.array([0, 1, 1, 0, 0])  # Binary data

# Calculate distances and similarities
euclidean_dist = euclidean(x, y)  # Euclidean distance
manhattan_dist = cityblock(x, y)  # Manhattan distance
cosine_sim = 1 - cosine(x, y)     # Cosine similarity (scipy's cosine returns distance, 1 - cosine is similarity)
jaccard_sim = jaccard_score(set_x, set_y)  # Jaccard similarity coefficient

# Print results
print(f"Euclidean Distance: {euclidean_dist:.4f}")
print(f"Manhattan Distance: {manhattan_dist:.4f}")
print(f"Cosine Similarity: {cosine_sim:.4f}")
print(f"Jaccard Similarity Coefficient: {jaccard_sim:.4f}")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Visualize Euclidean and Manhattan distances
axes[0].plot([x[0], y[0]], [x[1], y[1]], 'ro-', label='Euclidean Distance')
axes[0].plot([x[0], y[0]], [x[1], x[1]], 'g--', label='Manhattan Distance (X-axis)')
axes[0].plot([y[0], y[0]], [x[1], y[1]], 'b--', label='Manhattan Distance (Y-axis)')
axes[0].set_title(f"Euclidean vs Manhattan Distance\nEuclidean: {euclidean_dist:.2f}, Manhattan: {manhattan_dist:.2f}")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].legend()
axes[0].grid(True)

# Visualize Cosine Similarity
axes[1].arrow(0, 0, x[0], x[1], head_width=0.2, head_length=0.2, fc='r', ec='r', label='Vector x')
axes[1].arrow(0, 0, y[0], y[1], head_width=0.2, head_length=0.2, fc='b', ec='b', label='Vector y')
axes[1].set_title(f"Cosine Similarity: {cosine_sim:.2f}")
axes[1].set_xlabel("X-axis")
axes[1].set_ylabel("Y-axis")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# Visualize Jaccard Similarity Coefficient
plt.figure(figsize=(6, 4))
plt.bar(['Jaccard Similarity'], [jaccard_sim], color='purple')
plt.title(f"Jaccard Similarity Coefficient: {jaccard_sim:.2f}")
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```
